```python
import pandas as pd

books = pd.read_csv("./books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(),1,0)
books["age_of_the_book"] = 2025 - books["published_year"]

books
```

```python
columns_of_interest = ["num_pages", "average_rating", "age_of_the_book", "missing_description"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))

heatmap = sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',cbar_kws={"label" : "Correlation"})

plt.title("Correlation Matrix")
plt.show()
```

```python
books_missing = books[~books["description"].isna()&
     ~books["num_pages"].isna()&
     ~books["average_rating"].isna()&
     ~books["published_year"].isna()]

books_missing
```

```python
books_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
books_missing
```

```python
books_missing['words_in_description'] = books_missing['description'].str.split().str.len()
books_missing
```

```python
small = books_missing.loc[books_missing['words_in_description'].between(1,4) ,'description']
small
```

```python
books_missing.loc[books_missing['words_in_description'].between(15,24) ,'description']
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtile"] = (
    np.where(books_missing_25_words["subtitle"].isna(),books_missing_25_words["title"],books_missing_25_words[["title","subtitle"]].astype(str).agg(': '.join, axis=1))
) 
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13" , "description"]].astype(str).agg(': '.join, axis=1)
books_missing_25_words
```

```python
(
    books_missing_25_words.drop(
        ["subtitle",'missing_description',"words_in_description","age_of_the_book"],axis=1
    ).to_csv("books_cleaned.csv",index=False)
)
```
